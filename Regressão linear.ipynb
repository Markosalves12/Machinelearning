```python
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

# Base saude

```python
caminho_base_saude = r"E:\Data sets\curso de machine learning\plano_saude.csv".replace("\\","/")
```

```python
base_plano_saude = pd.read_csv(caminho_base_saude)
```

```python
base_plano_saude
```

```python
x_plano_saude = base_plano_saude.iloc[:,0].values
y_plano_saude = base_plano_saude.iloc[:,1].values
```

```python
# Correlação entre as variáveis
np.corrcoef(x_plano_saude, y_plano_saude)
```

```python
x_plano_saude = x_plano_saude.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
regressor_plano_saude = LinearRegression()
regressor_plano_saude.fit(x_plano_saude, y_plano_saude)
```

```python
print(regressor_plano_saude.intercept_)
print(regressor_plano_saude.coef_)
```

```python
previsoes = regressor_plano_saude.predict(x_plano_saude)
```

```python
grafico = px.scatter(x=x_plano_saude.ravel(), y=y_plano_saude)
grafico.add_scatter(x=x_plano_saude.ravel(), y=previsoes)
grafico.show()
```

```python
regressor_plano_saude.score(x_plano_saude, y_plano_saude)
```

```python
from yellowbrick.regressor import ResidualsPlot
visulizador = ResidualsPlot(regressor_plano_saude)
visulizador.fit(x_plano_saude, y_plano_saude)
visulizador.poof()
```

# Base casas

```python
caminho_base_casas = r"E:\Data sets\curso de machine learning\house_prices.csv".replace("\\","/")
```

```python
base_casas = pd.read_csv(caminho_base_casas)
```

```python
x_casas = base_casas.iloc[:, 5:6].values
y_casas = base_casas.iloc[:, 2].values
```

```python
from sklearn.model_selection import train_test_split
x_casas_treinamento, x_casas_teste, y_casas_treinamento, y_casas_teste = train_test_split(x_casas, y_casas, test_size=0.3, random_state=0)
```

```python
print(x_casas_treinamento.shape)
print(x_casas_teste.shape)
```

```python
print(y_casas_treinamento.shape)
print(y_casas_teste.shape)
```

```python
regressor_simples_casa = LinearRegression()
regressor_simples_casa.fit(x_casas_treinamento, y_casas_treinamento)
```

```python
print(regressor_simples_casa.intercept_)
print(regressor_simples_casa.coef_)
```

```python
print(regressor_simples_casa.score(x_casas_treinamento, y_casas_treinamento))
print(regressor_simples_casa.score(x_casas_teste, y_casas_teste))
```

```python
previsoes =  regressor_simples_casa.predict(x_casas_treinamento)
```

```python
previsoes_testes_casas =  regressor_simples_casa.predict(x_casas_teste)
```

```python
grafico = px.scatter(x=x_casas_treinamento.ravel(), y=previsoes)
# grafico.add_scatter(x=x_plano_saude.ravel(), y=previsoes)
grafico.show()
```

```python
grafico1 = px.scatter(x=x_casas_treinamento.ravel(), y=y_casas_treinamento)
grafico2 = px.line(x=x_casas_treinamento.ravel(), y=previsoes)
grafico2.data[0].line.color = 'red'
grafico3 = go.Figure(data=grafico1.data+grafico2.data)
grafico3.show()
```

### metricas de avaliação base casas

```python
mean_absolute_error(y_casas_teste, previsoes_testes_casas)
```

```python
mean_squared_error(y_casas_teste, previsoes_testes_casas)
```

```python
np.sqrt(mean_squared_error(y_casas_teste, previsoes_testes_casas))
```

# Regressão linear multipla na base casas

```python
x_casas_RM = base_casas.iloc[:,3:19].values
y_casas_RM = base_casas.iloc[:,2].values
```

```python
x_casas_treinamentoRM, x_casas_testeRM, y_casas_treinamentoRM, y_casas_testeRM = train_test_split(x_casas_RM, y_casas_RM, test_size=0.3, random_state=0)
```

```python
print(x_casas_treinamentoRM.shape, x_casas_testeRM.shape)
```

```python
print(y_casas_treinamentoRM.shape, y_casas_testeRM.shape)
```

```python
regressor_multiplo_casa = LinearRegression()
regressor_multiplo_casa.fit(x_casas_treinamentoRM, y_casas_treinamentoRM)
```

```python
print(regressor_multiplo_casa.intercept_)
print(regressor_multiplo_casa.coef_)
```

```python
print(regressor_multiplo_casa.score(x_casas_treinamentoRM, y_casas_treinamentoRM))
print(regressor_multiplo_casa.score(x_casas_testeRM, y_casas_testeRM))
```

```python
previsoesRMtestes = regressor_multiplo_casa.predict(x_casas_testeRM)
```

```python
print(mean_absolute_error(y_casas_testeRM, previsoesRMtestes))
print(mean_squared_error(y_casas_testeRM, previsoesRMtestes))
print(np.sqrt(mean_squared_error(y_casas_testeRM, previsoesRMtestes))) 
```
