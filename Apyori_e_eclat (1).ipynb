```python
from apyori import apriori
import pandas as pd
```

```python
base_mercado_1 = pd.read_csv("mercado.csv", header=None)
```

```python
base_mercado_1
```

```python
transacoes = list()
for i in range(len(base_mercado_1)):
  #print(i)
  transacoes.append([str(base_mercado_1.values[i, j]) for j in range(base_mercado_1.shape[1])])
```

```python
type(transacoes)
```

```python
transacoes
```

```python
regras = apriori(transacoes, min_support=0.3, min_lift=2, min_confidence=0.8)
resultados = list(regras)
```

```python
# construção e uma codificação para criar um data frame

SE = list()
ENTAO = list()
SUPPORT = list()
CONFIANCA =list()
LIFT = list()

for resultado in resultados:
  # print(resultado)
  s = resultado[1]
  # print(s)
  result_rules = resultado[2]
  # print(result_rule)
  for result_rule in result_rules:
    # print(result_rule)
    se = list(result_rule[0])
    entao = list(result_rule[1])
    confianca = result_rule[2]
    lift = result_rule[3]
    # print(se, '--', entao, '--', confianca, '--', lift)
    SE.append(se)
    ENTAO.append(entao)
    SUPPORT.append(s)
    CONFIANCA.append(confianca)
    LIFT.append(lift)
```

```python
rules_df = pd.DataFrame({
    'SE': SE,
    'ENTÃO': ENTAO,
    'SUPPORTE': SUPPORT,
    'CONFIANCA': CONFIANCA,
    'LIFT': LIFT
})
```

```python
rules_df.sort_values(['LIFT'], ascending=False)
```

```python
base_mercado_2 = pd.read_csv('mercado2.csv', header=None)
base_mercado_2
```

```python
transacoes2 = list()
for i in range(base_mercado_2.shape[0]):
  #print(i)
  transacoes2.append([str(base_mercado_2.values[i, j]) for j in range(base_mercado_2.shape[1])])
```

```python
transacoes2
```

```python
# Produtos vendidos 4 vezes por dia

(4*7)/7501 #4 vezes, vezes 7 dias da semana, dividido pelo numero de linhas
```

```python
regras2 = apriori(transacoes2, min_support=0.003, min_confidence=0.2, min_lift=3)
resultados2 = list(regras2)
len(resultados2)
```

```python
# construção e uma codificação para criar um data frame

SE2 = list()
ENTAO2 = list()
SUPPORT2 = list()
CONFIANCA2 =list()
LIFT2 = list()

for resultado2 in resultados2:
  # print(resultado)
  s2 = resultado2[1]
  # print(s)
  result_rules2 = resultado2[2]
  # print(result_rule)
  for result_rule2 in result_rules2:
    # print(result_rule)
    se2 = list(result_rule2[0])
    entao2 = list(result_rule2[1])
    confianca2 = result_rule2[2]
    lift2 = result_rule2[3]
    # print(se, '--', entao, '--', confianca, '--', lift)
    SE2.append(se2)
    ENTAO2.append(entao2)
    SUPPORT2.append(s2)
    CONFIANCA2.append(confianca2)
    LIFT2.append(lift2)
```

```python
rules_df2 = pd.DataFrame({
    'SE': SE2,
    'ENTÃO': ENTAO2,
    'SUPPORTE': SUPPORT2,
    'CONFIANCA': CONFIANCA2,
    'LIFT': LIFT2
})
```

```python
rules_df2.sort_values(['LIFT'], ascending=False)
```
